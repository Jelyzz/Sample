```python
import speech_recognition as sr
import pyttsx3

engine = pyttsx3.init()

def speech_to_text():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        try:
            audio = recognizer.listen(source)
            print("Recognizing...")
            text = recognizer.recognize_google(audio)
            print(f"Recognized Text: {text}")
            return text
        except sr.UnknownValueError:
            print("Sorry, I could not understand the audio.")
        except sr.RequestError as e:
            print(f"Could not request results from Google Speech Recognition service; {e}")
        return None

if __name__ == "__main__":
    recognized_text = speech_to_text()
    if recognized_text:
        print(f"You said: {recognized_text}")
        text_to_speech(f"You said: {recognized_text}")
```

```python
import pyttsx3

engine = pyttsx3.init()

def text_to_speech(text):
    engine.say(text)
    engine.runAndWait()

def get_text_input():
    text = input("Please enter the text: ")
    return text

# Example Usage:
if __name__ == "__main__":
    user_text = get_text_input()
    if user_text:
        print(f"You entered: {user_text}")
        text_to_speech(f"You entered: {user_text}")
```

```python
import speech_recognition as sr
import pyttsx3
import webbrowser

engine = pyttsx3.init()

def text_to_speech(text):
    engine.say(text)
    engine.runAndWait()

def speech_to_text():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        try:
            audio = recognizer.listen(source)
            print("Recognizing...")
            text = recognizer.recognize_google(audio)
            print(f"You said: {text}")
            return text.lower()
        except sr.UnknownValueError:
            print("Sorry, I could not understand the audio.")
            return None
        except sr.RequestError as e:
            print(f"Could not request results; {e}")
            return None

def execute_command(command):
    if "open youtube" in command:
        text_to_speech("Opening YouTube")
        webbrowser.open("https://www.youtube.com")
    elif "open google" in command:
        text_to_speech("Opening Google")
        webbrowser.open("https://www.google.com")
    else:
        text_to_speech("Sorry, I don't recognize that command.")

if __name__ == "__main__":
    text_to_speech("I am ready for your command.")
    command = speech_to_text()
    if command:
        execute_command(command)
```
